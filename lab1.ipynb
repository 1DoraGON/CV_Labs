## RGB Channels

– RGB channels are the essential components that collectively create the full spectrum of colors in digital images.

```python
R = img[:,:,0]
G = img[:,:,1]
B = img[:,:,2]
output = [img,R,G,B]
titles = ['Original Image', "R Channel", "G Channel", "B Channel"]
for i in range(4):
    plt.subplot(2, 2, i+1)
    plt.axis('off')
    plt.title(titles[i])
    if i==0:
        plt.imshow(output[i])
    else:
        plt.imshow(output[i], cmap='gray' )

plt.show()
```

– Stacking them together

```python
Output = np.dstack((R,G,B))
plt.imshow(Output)
plt.show()
```

### Load and plot the two images named (t4,t5) using matplotlib

### Increase the brightness of the pixels, making the image overall brighter

```python
darkFlowerImg = img + 0.5
plt.axis('off')
plt.imshow(darkFlowerImg)
plt.show()
```

### Reduce and Increase the difference between light and dark areas

```python
flowerImgHighContast = img.copy()
flowerImgLowContast = img.copy()
d = 0.1
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        for k in range(img.shape[2]):
            if (img[i, j, 0]+img[i, j, 1]+img[i, j, 2])/3 < 0.5:
                flowerImgHighContast[i,j,k] += d
                flowerImgLowContast[i,j,k] -= d
            else:
                flowerImgHighContast[i,j,k] -= d
                flowerImgLowContast[i,j,k] += d
plt.axis('off')
plt.imshow(flowerImgHighContast)
plt.show()
plt.axis('off')
plt.imshow(flowerImgLowContast)
plt.show()
```
